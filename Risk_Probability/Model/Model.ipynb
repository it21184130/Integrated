```python
import pandas as pd
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
Train_data = pd.read_csv("../New_Train.csv")
Test_data = pd.read_csv("../New_Test.csv")
```

```python
Train_data.shape
```

```python
Test_data.shape
```

```python
x_Test=Test_data.drop('risk_probability',axis=1)
y_Test=Test_data['risk_probability']
```

```python
x_Train=Train_data.drop('risk_probability',axis=1)
y_Train=Train_data['risk_probability']
```

```python
# !pip install sklearn
```

```python
x_Test.shape, x_Train.shape
```

```python
def model_acc(model):
    model.fit(x_Train,y_Train)
    acc=model.score(x_Test,y_Test)
    print(str(model)+ ' --> ' +str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr)

from sklearn.linear_model import Lasso
lasso = Lasso()
model_acc(lasso)

from sklearn.tree import DecisionTreeRegressor
Dt = DecisionTreeRegressor()
model_acc(Dt)

from sklearn.ensemble import RandomForestRegressor
Rf = RandomForestRegressor()
model_acc(Rf)
```

```python
model = RandomForestRegressor()
model.fit(x_Train, y_Train)

# Get predictions on the test dataset
# predictions = model.predict(x_Test)
```

```python
# Evaluate accuracy without cross-validation
train_accuracy = model.score(x_Train, y_Train)
test_accuracy = model.score(x_Test, y_Test)

print("Training Accuracy:", train_accuracy)
print("Test Accuracy:", test_accuracy)

model_acc(model)
```

```python
# from sklearn.model_selection import cross_val_score

# cv_scores = cross_val_score(model, x_Train, y_Train, cv=5)

# print("Cross-validation scores:", cv_scores)
# print("Average Cross-validation score:", cv_scores.mean())
```

```python
# Save the trained model using pickle
with open('best_model.pickle', 'wb') as file:
    pickle.dump(model, file)

print("Model saved successfully.")
```

```python
# Get predictions on the test dataset
predictions = model.predict(x_Test)
```

```python
def plot_actual_vs_predicted(y_test, predictions, title):
    plt.figure(figsize=(10, 6))
    plt.scatter(range(len(y_test)), y_test, color="blue", label="Actual")
    plt.scatter(range(len(predictions)), predictions, color="red", alpha=0.5, label="Predicted")
    plt.title(title)
    plt.xlabel("Samples")
    plt.ylabel("Risk Probability")
    plt.legend()
    plt.show()
```

```python
plot_actual_vs_predicted(
    y_Test, predictions,title="Actual vs Predicted Risk Probability"
)
```

```python
def risk_level(input_data):
    # Reshape data for model input
    input_data = np.array(input_data).reshape(1, -1)
    
    probabilities = model.predict(input_data)
    
    # Map probability to risk levels
    if probabilities <= 0.3:
        risk = 'Low Risk'
    elif 0.3 < probabilities <= 0.6:
        risk = 'Moderate Risk'
    else:
        risk = 'High Risk'
    
    return risk
```
